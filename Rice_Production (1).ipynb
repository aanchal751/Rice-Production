```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
dataset = pd.read_csv("/content/drive/MyDrive/Crops_data.csv")
print('Loading dataset from csv file and printing the head...\n')
dataset.head()
```

```python
dataset.info()
dataset_n = dataset
```

```python
dataset_n.drop(dataset_n.iloc[: , 8: ], inplace= True, axis=1)
dataset_n.head()
```

```python
dataset_n = dataset_n.drop(['Dist Code','State Code','Dist Name'], axis=1)
dataset_n.head()
```

```python
cols = list(dataset_n.columns)
a, b = cols.index('RICE PRODUCTION (1000 tons)'), cols.index('RICE YIELD (Kg per ha)')
cols[b], cols[a] = cols[a], cols[b]
dataset_n = dataset_n[cols]
dataset_n.head()
```

```python
plt.figure(figsize=(13, 5))
rice_production_by_year = dataset_n.groupby('Year')['RICE PRODUCTION (1000 tons)'].sum()
plt.plot(rice_production_by_year.index, rice_production_by_year.values, marker='o', color='lightgreen')
plt.title('Rice_Production (1000 tons) Over the Years', fontsize = 15)
plt.xlabel('Year', fontsize = 12)
plt.ylabel('Rice_Production', fontsize = 12)
plt.grid(True, linestyle = '--', alpha = 1.0)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(13, 5))
plt.bar(dataset_n.loc[:,"State Name"], dataset_n.loc[:,"RICE PRODUCTION (1000 tons)"], width=0.5, color = 'skyblue')
plt.title('Rice Production (1000 tons) by State', fontsize=14)
plt.xlabel('State', fontsize=12)
plt.ylabel('Rice Production (1000 tons)', fontsize=12)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
X = dataset_n.iloc[:, :-1].values
Y = dataset_n.iloc[:, -1].values
print(X)
print(Y)
```

```python
df_trial_A = pd.DataFrame(X, columns=['Year', 'State','Rice Area', 'Rice Yield'])
df_trial_A
```

```python
df_trial_B = pd.DataFrame(Y, columns=['Rice_Production'])
df_trial_B
```

```python
from sklearn.preprocessing import OneHotEncoder
ohe = OneHotEncoder(sparse_output=False).set_output(transform="pandas")
afohe = ohe.fit_transform(df_trial_A[['State']])
afohe
```

```python
df_trial_A = pd.concat([df_trial_A, afohe], axis= 1).drop(columns = ['State'])
df_trial_A
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(df_trial_A, df_trial_B, test_size= 0.2, random_state=0)
print(x_train)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x_train=sc.fit_transform(x_train)
x_test=sc.transform(x_test)
```

```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(x_train, y_train)
```

```python
y_pred = regressor.predict(x_test)
y_test.reset_index(drop=True)
y_test
```

```python
df_pred = pd.DataFrame(y_pred)
df_pred
```

```python
df_pred = df_pred.rename(columns = {0:"pred"})
df_test = y_test
df_test
```

```python
pd.concat([df_test,df_pred], axis=1)
```
